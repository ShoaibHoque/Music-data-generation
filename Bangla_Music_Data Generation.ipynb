# Getting the features of all tracks 

Here I'll get the feature of all tracks like how much energy, liveness, loudness, popularity each song has.

```python
def audio_features(album):
    #Add new key-values to store audio features
    spotify_albums[album]['acousticness'] = []
    spotify_albums[album]['danceability'] = []
    spotify_albums[album]['energy'] = []
    spotify_albums[album]['instrumentalness'] = []
    spotify_albums[album]['liveness'] = []
    spotify_albums[album]['loudness'] = []
    spotify_albums[album]['speechiness'] = []
    spotify_albums[album]['tempo'] = []
    spotify_albums[album]['valence'] = []
    spotify_albums[album]['popularity'] = []
    
    track_count = 0
    for track in spotify_albums[album]['uri']:
        #pull audio features per track
        features = sp.audio_features(track)
        
        #Append to relevant key-value
        spotify_albums[album]['acousticness'].append(features[0]['acousticness'])
        spotify_albums[album]['danceability'].append(features[0]['danceability'])
        spotify_albums[album]['energy'].append(features[0]['energy'])
        spotify_albums[album]['instrumentalness'].append(features[0]['instrumentalness'])
        spotify_albums[album]['liveness'].append(features[0]['liveness'])
        spotify_albums[album]['loudness'].append(features[0]['loudness'])
        spotify_albums[album]['speechiness'].append(features[0]['speechiness'])
        spotify_albums[album]['tempo'].append(features[0]['tempo'])
        spotify_albums[album]['valence'].append(features[0]['valence'])
        #popularity is stored elsewhere
        pop = sp.track(track)
        spotify_albums[album]['popularity'].append(pop['popularity'])
        track_count+=1
```

# Importing time library

The very reason to to use this time method just for not conflicting during fetching all tracks from each album, which I've faced before when I wanted to fech 10,000 songs from a playlist. The main reason was that tried to access all at once which can lead to an endless loop by trying to accessing all tracks together but the limit to get request from spotify is 400. To avoid those problem I am using this library.

```python
import time
import numpy as np
sleep_min = 2
sleep_max = 5
start_time = time.time()
request_count = 0
for i in spotify_albums:
    audio_features(i)
    request_count+=1
    if request_count % 5 == 0:
        print(str(request_count) + " playlists completed")
        time.sleep(np.random.uniform(sleep_min, sleep_max))
        print('Loop #: {}'.format(request_count))
        print('Elapsed Time: {} seconds'.format(time.time() - start_time))
```

# Store the artist information of each track

To store the information of each artist of each tracks, we have to make another dataframe individually which can concatinate with previous dataframe later

```python
artist_info = {}  # Dictionary to store artist information

for name in names:
    result = sp.search(q=name, type='artist', limit=1)
    artist = result['artists']['items'][0]

    artist_name = artist['name']

    if artist_name not in artist_info:
        artist_info[artist_name] = {
            'artist_name': artist_name,
            'artist_genres': artist['genres'],
            'artist_popularity': artist['popularity'],
            'num_album_tracks': 0,
            'num_single_tracks': 0,
            'num_total_tracks': 0
        }

    

    artist_albums = sp.artist_albums(artist['uri'], album_type='album,single')
    for album in artist_albums['items']:
        if album['album_type'] == 'album':
            artist_info[artist_name]['num_album_tracks'] += album['total_tracks']
        else:
            artist_info[artist_name]['num_single_tracks'] += album['total_tracks']

    artist_info[artist_name]['num_total_tracks'] = artist_info[artist_name]['num_single_tracks'] + artist_info[artist_name]['num_album_tracks']

for artist_name, info in artist_info.items():
    print(f"Artist: {artist_name}")
    print(f"Number of Album Tracks: {info['num_album_tracks']}")
    print(f"Number of Single Tracks: {info['num_single_tracks']}")
    print(f"Number of Total Tracks: {info['num_total_tracks']}")
    print("=" * 20)
```

```python
extended_artist_info = []
for artist_name, info in artist_info.items():
    num_tracks = info['num_album_tracks'] + info['num_single_tracks']
    artist_tracks = [info.copy() for _ in range(num_tracks)]
    extended_artist_info.extend(artist_tracks)

for entry in extended_artist_info:
    print(entry)

  
```

# Remove Duplicates

When the song is belongs to two artist then the that song uri id can be given to two artist together as both artist deserves the same song.
